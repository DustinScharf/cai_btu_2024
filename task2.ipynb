```python
# imports
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
from google.colab import drive
```

```python
# mounting the drive
drive.mount('/content/drive')
```

```python
# loading sample text from the file
with open("JaneEyre.txt", "r") as file:
  text = file.read()
```

```python
# building cloud: 8 minimal font size and black back
wordcloud = WordCloud(
  width=800,
  height=400,
  background_color="black",
  stopwords=set(STOPWORDS),
  min_font_size=8,
).generate(text)

# Plotting using plt
plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```

```python
# building cloud: 5 minimal font size and black back and custom stopwords
wordcloud = WordCloud(
  width=800,
  height=400,
  background_color="black",
  stopwords=set(["the", "a", "an"]),
  min_font_size=5,
).generate(text)

# Plotting using plt
plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```

```python
# building cloud: 21 maximal font size and pink back
wordcloud = WordCloud(
  width=1600,
  height=800,
  background_color="pink",
  stopwords=[],
  max_font_size=21,
).generate(text)

# Plotting using plt
plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```
